# Import Libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Import dataset

```
dataset=pd.read_csv('breast_cancer.csv')
x=dataset.iloc[:,1:-1].values
y=dataset.iloc[:,-1].values
```

```
print(x)
```

```
print(y)
```

# Split the data into training set and test set

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```
print(x_train)
```

```
print(y_train)
```

```
print(x_test)
```

```
print(y_test)
```

# Feature Scalling

```
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```
print(x_train)
```

```
print(x_test)
```

# Train the model on the Logistic Regression

```
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression(random_state=0)
clf.fit(x_train,y_train)
```

# Predict test set result

```
y_pred=clf.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# Making the confusion matrix

```
from sklearn.metrics import confusion_matrix,accuracy_score
print(confusion_matrix(y_test,y_pred))
accuracy_score(y_test,y_pred)
```

# Visualising the training set

```
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(x_train), y_train
X1, X2,X3,X4,X5,X6,X7,X8,X9 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 10, stop = X_set[:, 1].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 2].min() - 10, stop = X_set[:, 2].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 3].min() - 10, stop = X_set[:, 3].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 4].min() - 10, stop = X_set[:, 4].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 5].min() - 10, stop = X_set[:, 5].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 6].min() - 10, stop = X_set[:, 6].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 7].min() - 10, stop = X_set[:, 7].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 8].min() - 10, stop = X_set[:, 8].max() + 10, step = 0.25))
plt.contourf(X1, X2,X3,X4,X5,X6,X7,X8,X9, clf.predict(sc.transform(np.array([X1.ravel(), X2.ravel(),X3.ravel(),X4.ravel(),X5.ravel(),X6.ravel(),X7.ravel(),X8.ravel(),X9.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('yellow', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
plt.ylim(X3.min(), X3.max())
plt.ylim(X4.min(), X4.max())
plt.ylim(X5.min(), X5.max())
plt.ylim(X6.min(), X6.max())
plt.ylim(X7.min(), X7.max())
plt.ylim(X8.min(), X8.max())
plt.ylim(X9.min(), X9.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Brest Cancer)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

# Visualising the test set

```
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(x_test), y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, clf.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('yellow', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Brest Cancer)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
