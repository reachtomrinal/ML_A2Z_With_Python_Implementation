# Import Libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Import dataset

```
dataset=pd.read_csv('Mall_Customers.csv')
x=dataset.iloc[:,[3,4]].values
```

```
print(x)
```

# Using the Dendrogram method to find optimal number of clusters

```
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.legend()
plt.show()
```

# Training the Hierarchial model on the dataset

```
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_hc=hc.fit_predict(x)
print(y_hc)
```

# Visualising the Clustures

```

for j in range(0,5):
  plt.scatter(x[y_hc == j,0],x[y_hc == j,1],s=100,c=str(np.random.random()),label='Cluster'+str(j))
#plt.scatter(hc.cluster_centers_[:,0],hc.cluster_centers_[:,1],s=300,c='yellow',label='Centroid')
plt.title('Cluster of customers')
plt.xlabel('Annual Income(k£)')
plt.ylabel('Spending Score(1-100)')
plt.legend()
plt.show()
```
