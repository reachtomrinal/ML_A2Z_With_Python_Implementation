# Import Libraries

```
import numpy as np
import matplotlib.pyplot as pt
import pandas as pd
```

# importing datasheet

```
datasheet = pd.read_excel('TableLibrary.xlsx')
X = datasheet.iloc[:,:-1].values
Y = datasheet.iloc[:,-1].values
```

```
print(X)
```

```
print(Y)
```

# Taking care of Missing Data

```
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:,1:])
X[:,1:] = imputer.transform(X[:,1:])
```

```
print(X)
```

# Encoding Catagory












## Encoding Independent Variable

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:,0] = le.fit_transform(X[:,0])
```

```
print(X)
```

## Encoding dependent variable.
We do not want to do any encoding for the dependent vafriable

# Spliting the data into training set and test set

```
from sklearn.model_selection import train_test_split
X_Tran, X_Test, Y_Tran, Y_Test=train_test_split(X,Y,test_size=0.2,random_state=1)
```
