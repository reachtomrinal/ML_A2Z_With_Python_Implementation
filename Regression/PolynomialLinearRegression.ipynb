# Import Libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Import dataset

```
dataset=pd.read_csv('Position_Salaries.csv')
x=dataset.iloc[:,1:-1].values
y=dataset.iloc[:,-1].values
```

```
print(x)
```

```
print(y)
```

# Training the linear regression model on the whole dataset

```
from sklearn.linear_model import LinearRegression
lin_Reg=LinearRegression()
lin_Reg.fit(x,y)
```

# Training the Polynomial regression model on the whole dataset

```
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
x_poly=poly_reg.fit_transform(x)
lin_Reg_2=LinearRegression()
lin_Reg_2.fit(x_poly,y)
```

# Visualising Linear Regression Result

```
plt.scatter(x,y,color='Red')
plt.plot(x,lin_Reg.predict(x),color='blue')
plt.title('Salary Predection')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

# Visualising Polynomial Regression Result

```
plt.scatter(x,y,color='Red')
plt.plot(x,lin_Reg_2.predict(x_poly),color='blue')
plt.title('Salary Predection')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

# Predecting a new result with Linear Regression

```
lin_Reg.predict([[6.5]])
```

# Predecting a new Result with the polynomial regression

```
lin_Reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
